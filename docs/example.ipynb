# Example usage

To use `simplecalcpypip.simplecalcpypip` module in a project:

```python
import simplecalcpypip.simplecalcpypip as spclc

dir(spclc)
```

### Função soma(a, b)

```python
resultado = spclc.subtracao(10, 4)
print(resultado)  # Saída: 6
```


To use `simplecalcpypip.simpleplot` module in a project:

```python
import simplecalcpypip.simpleplot as spplt

dir(spplt)
```

### Função plot_grafico_linear(x, y)

```python
x_linear = spplt.np.linspace(0, 10, 100)
y_linear = 2 * x_linear + 5
spplt.plot_grafico_linear(x_linear, y_linear, "X", "Y", "Gráfico Linear")
```
